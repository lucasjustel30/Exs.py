Enter a number and print all its divisors 

```python

num = int(input('Enter a number to calculate all its divisors'))

div = 1

while div < num:
    if num%div == 0:
      print(div,'div is divisor of',num)
    div = div + 1




```

Enter a sequence of numbers to sum and press zero to finish

```python

num = int(input('enter a number to sum and press zero to stop'))

sum = 0

while num != 0:
    sum = sum + num
    num = int(input('enter a number to sum and press zero to stop'))

print('the sum is:' , sum)




```
